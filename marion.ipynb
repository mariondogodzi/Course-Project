```python
#Histograms for numeric columns
for column in dataset.columns:
    if dataset[column].dtype in ['int64', 'float64']:
        mpl.hist(dataset[column], bins=20, color='blue', alpha=0.7)
        mpl.xlabel('X-axis label')
        mpl.ylabel('Y-axis label')
        mpl.title(f'Histogram of {column}')
        mpl.show()
```

```python
# Create scatter plots for pairs of numeric columns
numeric_columns = dataset.select_dtypes(include=['int64', 'float64']).columns

for i in range(len(numeric_columns)):
    for j in range(i+1, len(numeric_columns)):
        mpl.scatter(dataset[numeric_columns[i]], dataset[numeric_columns[j]])
        mpl.xlabel(numeric_columns[i])
        mpl.ylabel(numeric_columns[j])
        mpl.title(f'Scatter Plot: {numeric_columns[i]} vs {numeric_columns[j]}')
        mpl.show()
```

```python
# Correlation heatmap 
correlation_matrix = dataset.corr()

mpl.figure(figsize=(10, 8))
mpl.imshow(correlation_matrix, cmap='coolwarm', interpolation='none')
mpl.colorbar().set_label('Correlation Coefficient')
mpl.xticks(range(len(correlation_matrix)), correlation_matrix.columns, rotation=90)
mpl.yticks(range(len(correlation_matrix)), correlation_matrix.columns)
mpl.title('Correlation Heatmap')
mpl.show()

```
